```python
import numpy as np
co2 = np.loadtxt("mlo_co2.txt", comments="#", usecols=(3,), unpack=False) #skiprows=4, 
yearmonth = np.loadtxt("mlo_co2.txt", comments="#", usecols=(1,2), dtype=int, unpack=True)  #skiprows=4, 
```

##### What did usecols=(1,2,3) do? Numpy.loadtxt: http://docs.scipy.org/doc/numpy/reference/generated/numpy.loadtxt.html

```python
co2 = co2[4:]
yearmonth = yearmonth[:,4:]
print co2.shape, yearmonth.shape
```

##### Now we have plotted the CO2 data. Can you convert the year and month coloumns into datetime format, and then lable the x-axis with the datetime?

```python
import datetime as DT
#assume all days are the first day of month
days = np.ones_like(co2, dtype=np.integer)
#year,doy = datetime.datetime(data[4:,0], data[4:,1], days).strftime('%Y %j').split()
#months = data[4:,1]

dates = []

for i in np.arange(len(days)):
    dates.append(DT.datetime(yearmonth[0][i], yearmonth[1][i], days[i]))
    
print dates[-1]
```

##### We can convert the 0st and 1st column of our data into datetime format, and use such datetime format to plot our time series data

```python
plt.figure(figsize=(20,5)) #make figure wider
plt.plot(dates,co2)
```

## 2. AR & MA

##### Now, let's explore this series

```python
import scipy.stats as stats

index = np.arange(len(co2))

co2_lin = stats.linregress(index, co2)
co2_predict = co2_lin[0] * index + co2_lin[1]

#figure(figsize=(20,5))
plot(dates, co2)
plot(dates, co2_predict)
title('Mauna Loa CO2 Concentation')
xlabel('Year')
ylabel('CO2 (ppm)')
figtext(0.15, 0.8, 'r = '+str(co2_lin[2]), size='x-large')
figtext(0.15, 0.75, 'intercept = '+str(co2_lin[1]), size='large')
figtext(0.15, 0.7, 'slope = '+str(co2_lin[0]), size='large')
```

##### R-square looks really high, but we can see with our eyes that the trend is not a straight line. Next, we example the residuals

```python
#calculate residuals
co2_lin_redis = co2 - co2_predict
figure(figsize=(20,5))
plot(dates, co2_lin_redis)
```

```python
import statsmodels.formula.api as smf
co2_model = smf.OLS(co2, index).fit()
print co2_model.summary()
co2_pr = co2_model.predict()
plot(dates, co2)
plot(dates, co2_pr)
```
